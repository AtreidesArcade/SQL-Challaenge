```python
#Dependencies 
#SQL Alchemy
from sqlalchemy import create_engine
from config import username, password
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Create connection to database
engine = create_engine(f'postgresql://{username}:{password}@localhost:5433/postgres')
connection = engine.connect()
```

```python
#Get records from Eomployees Table
employees_data = pd.read_sql("select * from employees", connection)
#Display Employees Table
employees_data.head()
```

```python
# Join the employees and salaries tables 
employee_salary_data = pd.read_sql('select employees.emp_no, salaries.salary from employees as employees inner join salaries as salaries on employees.emp_no = salaries.emp_no', connection)
#Display salary data
employee_salary_data.head()
```

```python
#Use a Histogram to visualize the most common salary ranges for employees. 
plt.figure(figsize=(15, 10))
plt.hist(employee_salary_data["salary"], color='#0172D8', bins=15)
plt.ylabel(f'Number of employees',fontsize=15)
plt.xlabel('Salary ($)',fontsize=15)
plt.title("Frequency Distribution of salary ranges for employees",fontsize=20)
plt.grid()
plt.savefig("./Images/employee_salary_distribution.png", bbox_inches='tight')
plt.show()
```

```python
#Select all records in the Title tables
titles = pd.read_sql("Select * From titles", connection)
titles.head()
```

```python
#Rename emp_title_id to match title tables
employees_titles_id = employees_data.rename(columns={"emp_title_id":"title_id"})
employees_titles_id.head()
```

```python
#combine employee and title tables
combined_data = pd.merge(employees_titles_id,titles, on='title_id', how='inner')
combined_data.head()
```

```python
#Combine employee , title, salary tables
merge_data = pd.merge(combined_data,employee_salary_data, on='emp_no', how="inner")
merge_data.head()
```

```python
#Create merged employee-title and salary table 
title_salary = merge_data[['title','salary']]
title_salary.head()
```

```python
#Group salary data by title
salary_by_title = title_salary.groupby(['title']).mean()
salary_by_title
```

```python
#Create new Index
index_avg_salary = salary_by_title.reset_index()
index_avg_salary
```

```python
# Set x_axis, y_axis & Tick Locations
fig1, ax1 = plt.subplots(figsize=(15, 10))
x_axis = index_avg_salary["title"]
ticks = np.arange(len(x_axis))
y_axis = index_avg_salary["salary"]



# Create Bar Chart Based on Above Data
plt.bar(x_axis, y_axis, align="center",alpha=0.8, color=["#000000", "#E69F00", "#56B4E9", "#86ADA2", "#0072B2", "#D55E00", "#CC79A7"])

# Create Ticks for Bar Chart's x_axis
plt.xticks(ticks, x_axis, rotation="vertical")

# Set Labels & Title
plt.ylabel("Salaries ($)",fontsize=15)
plt.xlabel("Employee Titles",fontsize=15)
plt.title("Average Employee Salary by Title",fontsize=20)

# Save Figure
plt.savefig("./Images/average_salary_by_title.png")

# Show plot
plt.show()
```
